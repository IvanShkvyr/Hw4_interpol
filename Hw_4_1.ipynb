```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
from scipy import integrate
```

```python
speed = np.array([25, 35, 45, 30, 60, 120, 100, 100, 70, 75, 80, 65])
time = np.linspace(0, 11, 12)
```

```python
time
```

```python
plt.figure(figsize=(12,8), dpi= 80)

plt.plot(time, speed)

plt.xlabel('Час, год', size="x-large")
plt.ylabel('Швидкість в моменті часу t, км/год', size="x-large")
plt.title('Графік зміни швидкості транспортного засобу з часом', fontsize=15)
plt.ylim(0, 130)
plt.xlim(1, 11)
plt.grid(alpha=0.5)
plt.show()
```

```python
f_cubic = interp1d(time, speed, kind='cubic')
```

```python
plt.figure(figsize=(12,8), dpi= 80)

more_velues = np.linspace(0, 11, 10000)
plt.plot(more_velues, f_cubic(more_velues), 'green')
plt.plot(time, speed, 'bo')
plt.xlabel('Час, год', size="x-large")
plt.ylabel('Швидкість в моменті часу t, км/год', size="x-large")
plt.title('Графік зміни швидкості транспортного засобу з часом (метод інтерполяції cubic)', fontsize=15)
plt.ylim(0, 130)
plt.xlim(1, 11)
plt.grid(alpha=0.5)
plt.show()
```

```python
result_cubic = integrate.quad(lambda x: f_cubic(x), 0, 11)
print(f"Шлях пройдений точкою при нерівномірному русі становить (метод "\
      f"інтерполяції cubic)- {round(result_cubic[0], 2)} км.")
```

```python
f_quadratic = interp1d(time, speed, kind='quadratic')
```

```python
plt.figure(figsize=(12,8), dpi= 80)

more_velues = np.linspace(0, 11, 10000)
plt.plot(more_velues, f_quadratic(more_velues), 'red')
plt.plot(time, speed, 'bo')
plt.xlabel('Час, год', size="x-large")
plt.ylabel('Швидкість в моменті часу t, км/год', size="x-large")
plt.title('Графік зміни швидкості транспортного засобу з часом (метод інтерполяції quadratic)', fontsize=15)
plt.ylim(0, 130)
plt.xlim(1, 11)
plt.grid(alpha=0.5)
plt.show()
```

```python
result_quadratic = integrate.quad(lambda x: f_quadratic(x), 0, 11)
print(f"Шлях пройдений точкою при нерівномірному русі становить (метод "\
      f"інтерполяції quadratic) - {round(result_quadratic[0], 2)} км.")
```

```python
print(f"Отримане значення довжини пройденого шляху за допомогою методу "\
      " інтерполяції quadratic, перевищує значення, отримане методом cubic на "\
      f"{round(result_quadratic[0] - result_cubic[0], 2)} км.")
```
